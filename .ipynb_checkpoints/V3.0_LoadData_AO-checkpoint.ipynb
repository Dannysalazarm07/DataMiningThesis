## Loading MATRIZ_MULTINIVEL.csv (All variables) and accidente_ofidico.csv (AO ocurrence variable)

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 500)
```

```python


accidente_ofidico_path = "datos/accidente_ofidico.csv"
ao_df = pd.read_csv(accidente_ofidico_path)
variables_path = "datos/PLANTILLA_MULTINIVEL.csv"
variables_df = pd.read_csv(variables_path)
del variables_df['esperados']
```

```python
#print(variables_df.head())
print(ao_df.head())
```

## Join ao_df to variables_df

```python
#selecting columns in data set with ao ocurrence in th whole period (2007 - 2016)
only_ocurrences = ao_df.iloc[:,20:140]
only_occurences_col = only_ocurrences.values.flatten()
variables_df['casos_'] = only_occurences_col
print(variables_df['casos_'])
variables_df.to_csv('pruebas.csv')
```

# Add total population

```python
#Anadir poblacio  total al data set de variables
only_population = ao_df.iloc[:,6:16]
dfNewPopulation = only_population.iloc[:,0]
columsNames = ["poblacion_2007", "poblacion_2008", "poblacion_2009", "poblacion_2010", "poblacion_2011", 
               "poblacion_2012", "poblacion_2013", "poblacion_2014", "poblacion_2015", "poblacion_2016"]

for i in range(len(columsNames)):
    thisColum = only_population.iloc[:,i]
    if i==0:
        d=11
    else:
        d=12
    for j in range(d):
        dfNewPopulation = pd.concat([dfNewPopulation, thisColum], axis=1)

dfNewPopulation.to_csv('total_poblacion.csv', encoding='latin-1')

#selecting columns in data set with ao ocurrence in th whole period (2007 - 2016)
only_population_col = dfNewPopulation.values.flatten()
variables_df['total_poblacion'] = only_population_col
variables_df.to_csv('total_variables_data_set.csv', encoding='latin-1')
```

# Adding tempProm and precProm columns

```python
# Abriir los dos archivos (multinivel y temp y prec promedios multianuales mensuales)
variables_path = "datos/MULTINIVEL_AO.csv"
variables_df = pd.read_csv(variables_path, encoding='latin-1')
temPrec_path = "datos/final_data_table_r.csv"
temPrec_df = pd.read_csv(temPrec_path, encoding='latin-1')
temPrec_df.head()
```

```python

temPrec_df.set_index('Codigo DANE-mes', inplace=True)
temPrec_df
```

```python
#Guarda columna danemes en variable index para luego agregarla al df y asiganarla como el indice del mismo
index = variables_df['danemes'].copy(deep = True)
variables_df['Index'] = index
variables_df.set_index('Index',inplace=True)
variables_df.index
```

```python
#Agrega las columnas promedio temperatura y precipitación de R al df multinivel
df = variables_df.merge(temPrec_df[['PromedioTemp','PromedioPrec']], left_index=True, right_index=True)
df.head()

```

## Code for testing mergin dataframes

```python
index_values = df.index
import random
samples_random = 2000
random_index_values = random.sample(list(index_values), 2000)
validation = True
validation_column = 'PromedioPrec'

for sample in index_values:
    df1_value = df[validation_column][sample]
    df2_value = temPrec_df[validation_column][sample]
    if not(np.isnan(df1_value) and np.isnan(df2_value)):
        if df1_value != df2_value:
            validation = False
            break
            
print(validation)
    
```
